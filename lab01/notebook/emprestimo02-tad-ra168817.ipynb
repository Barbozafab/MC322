# Transformando Empréstimo em um TAD

Considere a resolução para múltiplos empréstimos.

```c
#include <stdio.h>

typedef struct {
    float s;
    int   n;
    float j;
    int   corrente;
    float p;
} Emprestimo;

Emprestimo novoEmprestimo(float s, int n, float j) {
    Emprestimo umEmprestimo;
    umEmprestimo.s = s;
    umEmprestimo.n = n;
    umEmprestimo.j = j;
    umEmprestimo.corrente = 1;
    umEmprestimo.p = s;
    return umEmprestimo;
}

float proximaParcela(Emprestimo *umEmprestimo) {
    float retorno = umEmprestimo->p;
    if (umEmprestimo->corrente < umEmprestimo->n)
        umEmprestimo->p = umEmprestimo->p + (umEmprestimo->p * (umEmprestimo->j/100));
    else
        umEmprestimo->p = 0;
    (umEmprestimo->corrente)++;
    return retorno;
}

int main() {
    Emprestimo emprestimo1,
               emprestimo2;
    
    emprestimo1 = novoEmprestimo(200, 5, 1);
    emprestimo2 = novoEmprestimo(500, 7, 2);
    
    float p1 = proximaParcela(&emprestimo1),
          p2 = proximaParcela(&emprestimo2);
    
    int i = 1;
    while (p1 > 0 || p2 > 0) {
        if (p1 > 0)
            printf("Emprestimo 1: parcela %d eh %3.2f\n", i, p1);
        if (p2 > 0)
            printf("Emprestimo 2: parcela %d eh %3.2f\n", i, p2);
        p1 = proximaParcela(&emprestimo1);
        p2 = proximaParcela(&emprestimo2);
        i++;
    }   
}
```

# Exercício

Modifique este código (ou um código que você mesmo tenha feito) para que a estrutura (struct) possa representar um Tipo Abstrato de Dados (TAD), em que ela contenha não apenas os valores de estado,como também operações executadas sobre eles. As operações (funções em C) também devem fazer parte do struct do mesmo modo que os campos de dados.

```c
#include <stdio.h>

typedef struct Emprestimo emprestimo_t;

struct Emprestimo {
    double s;
    int n;
    int atual;
    int j;
    double (*prox_parcela)(emprestimo_t *);
};

double prox_parcela(emprestimo_t *);

emprestimo_t criar_emprestimo(double s, int n, int j) {
    emprestimo_t novo;
    novo.s = s;
    novo.n = n;
    novo.atual = 1;
    novo.j = j;
    novo.prox_parcela = prox_parcela;
    return novo;
}

double prox_parcela(emprestimo_t *emprestimo) {
    double parcela_atual = emprestimo->s;
    emprestimo->atual ++;
    if (emprestimo->atual > emprestimo->n) {
        emprestimo->s = 0;
    } else {
        emprestimo->s *= (1 + 0.01 * emprestimo->j);
    }
    return parcela_atual;
}

int main() {
    emprestimo_t emprestimo1 = criar_emprestimo(200, 5, 1);
    emprestimo_t emprestimo2 = criar_emprestimo(500, 7, 2);
    double parcela1 = emprestimo1.prox_parcela(&emprestimo1);
    double parcela2 = emprestimo2.prox_parcela(&emprestimo2);
    int i = 1;
    while (parcela1 > 0 || parcela2 > 0) {
        if (parcela1 > 0) {
            printf("Emprestimo 1: parcela %d eh %.2f\n", i, parcela1);
            parcela1 = emprestimo1.prox_parcela(&emprestimo1);
        }
        if (parcela2 > 0) {
            printf("Emprestimo 2: parcela %d eh %.2f\n", i, parcela2);
            parcela2 = emprestimo2.prox_parcela(&emprestimo2);
        }
        i ++;
    }
}
```
